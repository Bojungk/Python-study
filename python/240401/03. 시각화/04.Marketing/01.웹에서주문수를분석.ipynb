### 1 : 데이터 읽기

```python
# customer_master.csv : 쇼핑몰 사이트의 고객 정보
import pandas as pd

customer_master = pd.read_csv(r'C:\유비온_project1\03. 데이터분석\03. 시각화\04.Marketing\Data in\customer_master.csv')
customer_master.head()
```

```python
# item_master.csv : 상품데이터
item_master = pd.read_csv(r'C:\유비온_project1\03. 데이터분석\03. 시각화\04.Marketing\Data in\item_master.csv')
item_master.head()
```

```python
# transaction_1.csv와 transaction_2.csv : 구매 내역 데이터
transaction_1 = pd.read_csv(r'C:\유비온_project1\03. 데이터분석\03. 시각화\04.Marketing\Data in\transaction_1.csv')
transaction_2 = pd.read_csv(r'C:\유비온_project1\03. 데이터분석\03. 시각화\04.Marketing\Data in\transaction_2.csv')

transaction_1.head()
```

### 2 : 데이터 결합(유니언)

```python
# concat : 세로로 합치기
transaction = pd.concat([transaction_1, transaction_2], ignore_index=True)
transaction.head()
```

```python
# len() : 길이 확인 / 유니언 결과 확인
print(len(transaction_1))
print(len(transaction_2))
print(len(transaction))
```

### 3 : 매출 데이터끼리 결합(조인)

```python
transaction_detail_1 = pd.read_csv(r'C:\유비온_project1\03. 데이터분석\03. 시각화\04.Marketing\Data in\transaction_detail_1.csv')
transaction_detail_1.head()
```

```python
transaction_detail_2 = pd.read_csv(r'C:\유비온_project1\03. 데이터분석\03. 시각화\04.Marketing\Data in\transaction_detail_2.csv')
transaction_detail_2.head()
```

```python
transaction_detail=pd.concat([transaction_detail_1,transaction_detail_2], \
    ignore_index=True)
transaction_detail.head()
```

```python
join_data = pd.merge(transaction_detail, transaction[["transaction_id", \
    "payment_date", "customer_id"]], 
                    on="transaction_id", how="left")
join_data.head()
```

```python
print(len(transaction_detail))
print(len(transaction))
print(len(join_data))
```

### 4 : 마스터데이터를 결합(조인)

```python
# 고객정보, 상품정보 : 4종류의 6개의 데이터 결합
join_data = pd.merge(join_data, customer_master, on="customer_id", how="left")
join_data = pd.merge(join_data, item_master, on="item_id", how="left")
join_data.head()
```

### 5 : 필요한 데이터 컬럼 만들기

```python
# price 필드 추가
join_data["price"] = join_data["quantity"] * join_data["item_price"]
join_data[["quantity", "item_price","price"]].head()
```

### 8 : 월별로 데이터를 집계

```python
join_data.dtypes
```

```python
# payment_date를 datetime형으로 변환
# 구입일 : payment_date에서 연월을 추출 후 새로운 컬럼 추가
# 연월 단위로 price를 집계

join_data["payment_date"] = pd.to_datetime(join_data["payment_date"])
join_data["payment_month"] = join_data["payment_date"].dt.strftime("%Y%m")
join_data[["payment_date", "payment_month"]].head()
```

```python
# 월별 매출 집계
# 결과 : 5월의 매출이 조금 내려갔고, 6~7은 회복, 반년동안 가장 매출이 높은 달은 7월
join_data.groupby("payment_month")["price"].sum()
```

### 9 : 월별, 상품별 데이터를 집계

```python
# 월별, 상품별로 매출의 합계, 수량을 표시 
```

```python
monthly_sales = join_data.groupby(["payment_date", "payment_month"]).agg(
total_sales=('price', lambda x: sum(x)),
total_quantity=('quantity', 'sum')
).reset_index()
```

```python
monthly_sales.head()
```

```python
#결과: 매출의 합계 PC-E가 가장높음, 수량은 가격이 싼 PC-A가 높음 
#상품별 5월 PC-B, PC-D가 증가, 큰 매출을 차지하는 PC-E의 매출의 매출이 감소
#6월 7월은 PC-E의 매출이 크게 상승 
pd.pivot_table(join_data, index='item_name', columns='payment_month',\
    values=['price', 'quantity'], aggfunc='sum')
```

### 10 : 상품별 매출 추이를 가시화

```python
graph_data=pd.pivot_table(join_data, index='payment_month', \
                          columns='item_name', values='price', aggfunc='sum')
graph_data.head()
```

```python
import matplotlib.pyplot as plt

plt.plot(list(graph_data.index), graph_data["PC-A"], label='PC-A')
plt.plot(list(graph_data.index), graph_data["PC-B"], label='PC-B')
plt.plot(list(graph_data.index), graph_data["PC-C"], label='PC-C')
plt.plot(list(graph_data.index), graph_data["PC-D"], label='PC-D')
plt.plot(list(graph_data.index), graph_data["PC-E"], label='PC-E')
plt.legend()
```
