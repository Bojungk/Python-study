## Iframe popup

```python
from folium.map import Popup
import pandas as pd
import folium

df=pd.read_excel(r'C:\유비온_project1\03. 데이터분석\03. 시각화\data-in\서울지역 대학교 위치.xlsx')

seoul_map=folium.Map(location=[37.55,126.98], zoom_start=12)
```

```python
for name, lat, lng in zip(df.대학교, df.위도, df.경도):
    iframe = folium.IFrame(name, width=300, height=50)
    popup = folium.Popup(iframe, max_width = 300)
    folium.Marker([lat, lng], popup=popup).add_to(seoul_map)
   
seoul_map.save('./seoul_colleges.html')
seoul_map
```

### 원형 마커


```python
from folium.map import Popup
import pandas as pd
import folium

df=pd.read_excel(r'C:\유비온_project1\03. 데이터분석\03. 시각화\data-in\서울지역 대학교 위치.xlsx')

seoul_map=folium.Map(location=[37.55,126.98], zoom_start=12)
```

```python
for name, lat, lng in zip(df.대학교, df.위도, df.경도):
    iframe = folium.IFrame(name, width=300, height=50)
    popup = folium.Popup(iframe, max_width = 300)
    folium.CircleMarker([lat, lng],
                        radius=10,                           #원의 반지름 
                        color='brown',                       #원의 둘레 색상
                        fill=True,
                        fill_color='coral',                  #원을 채우는 색
                        fill_opacity=0.7,                    #투명도
                        popup=popup).add_to(seoul_map)
       
seoul_map.save(r'C:\유비온_project1\03. 데이터분석\03. 시각화\data-out\seoul.html')
seoul_map
```

### 경기도 인구 변화

```python
import pandas as pd
import folium
import json

file_path=r'C:\유비온_project1\03. 데이터분석\03. 시각화\data-in\경기도인구데이터.xlsx'
df= pd.read_excel(r'C:\유비온_project1\03. 데이터분석\03. 시각화\data-in\경기도인구데이터.xlsx'
    ,index_col='구분', engine='openpyxl')

df.columns=df.columns.map(str)
```

```python
#경기도 시군구 경계 정보를 가진 geo-json 파일 불러오기
geo_path= r'C:\유비온_project1\03. 데이터분석\03. 시각화\data-in\경기도행정구역경계.json'

try:
    geo_data= json.load(open(r'C:\유비온_project1\03. 데이터분석\03. 시각화\data-in\경기도행정구역경계.json', encoding= 'utf-8'))

except:
    geo_data= json.load(open(r'C:\유비온_project1\03. 데이터분석\03. 시각화\data-in\경기도행정구역경계.json', encoding= 'utf-8-sig'))
```

```python
# 경기도 지도 만들기
g_map = folium.Map(location=[37.5502,126.982], zoom_starts=9)
```

```python
year='2007' 
#Choropleth 클래스로 단계구분도 표시하기
folium.Choropleth(geo_data=geo_data,
                  data = df[year],
                  columns=[df.index, df[year]],
                  fill_color='YlOrRd', fill_opacity=0.7, line_opacity=0.3,
                  threshold_scale=[10000, 100000, 300000, 500000, 700000],
                  key_on='feature.properties.name',
).add_to(g_map)

g_map.save('./gyonggi_population'+ year + '.html')
```
